```python
import bs4, requests

url='https://google.com'
headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36',
    }
res = requests.get(url, headers=headers) # downloads the page
res.raise_for_status() # will throw error if not 200
```

```python
soup = bs4.BeautifulSoup(res.text, 'html.parser') # parse
```

```python
type(soup)
```

# Actions
`soup.prettify()` makes html readable

`print(soup.find_all('b')[0].attrs)` will return a dictionary of all attributes

`soup.find_all('b')[0].string` will give the actual text between the `<b>` tags

`soup.find_all('b')[0].string.replace_with('new stuff')` will change the string

# Properties available
## select
we can use `soup.select('.class')` or `soup.select('#id')` or others: https://automatetheboringstuff.com/chapter11/#calibre_link-2958

## find
`soup.find('p')` or `soup.p` will print out first `<p>` tag

`soup.p.name` gives name of first p tag (which is `p`) so you can change it to whatever name you want.

## findall
`links=soup.find_all('a')` gives a list of all `a`

`link.attrs['href']` gives the attribute

`soup.find_all('b')[0]['id']` actually returns the `id` attribute

## get
The `get()` method for Tag objects makes it simple to access attribute values from an element. The method is passed a string of an attribute name and returns that attribute’s value. Using example.html, enter the following into the interactive shell:

`soup = bs4.BeautifulSoup(open('example.html'))
spanElem = soup.select('span')[0]
str(spanElem)`

'<span id="author">Al Sweigart</span>'

`spanElem.get('id')`

'author'

`spanElem.get('some_nonexistent_addr') == None`

True

`spanElem.attrs`

{'id': 'author'}

Here we use select() to find any <span> elements and then store the first matched element in spanElem. Passing the attribute name 'id' to get() returns the attribute’s value, 'author'.
